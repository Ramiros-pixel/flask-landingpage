14.	Kelompokkan data berdasarkan tahun dan hitung total jumlah_penderita_dm per tahun di Jawa Barat.

```python
jumlah_penderita_pertahun = df.groupby('tahun')['jumlah_penderita_dm'].sum().reset_index()
jumlah_penderita_pertahun
```

15.	Kelompokkan data berdasarkan nama_kabupaten_kota dan hitung rata-rata jumlah_penderita_dm tiap kabupaten/kota (jika terdapat lebih dari satu tahun).

17.	Buat kolom baru bernama kategori_dm dengan aturan berbasis jumlah_penderita_dm: < 50.000 → "Rendah", 50.000–99.999 → "Sedang", >= 100.000 → "Tinggi"


```python
df['kategori_dm'] = pd.cut(
    df['jumlah_penderita_dm'],
    bins=[-1, 49999, 99999, float('inf')],
    labels=['Rendah', 'Sedang', 'Tinggi']
)
df
```

18.	Untuk setiap tahun, hitung persentase jumlah penderita DM per kabupaten/kota terhadap total jumlah penderita DM di tahun tersebut, lalu simpan dalam kolom baru bernama persentase_tahun.

```python
df['persentase_tahun'] = ( df['jumlah_penderita_dm'] / df.groupby('tahun')['jumlah_penderita_dm'].transform('sum')) * 100
df['persentase_tahun'] = df['persentase_tahun'].round(2)
df
```

19.	Buat tabel ringkas (misalnya menggunakan groupby) yang menunjukkan:
tahun, total_jumlah_penderita_dm, dan jumlah_kabupaten_kota yang datanya tercatat di tahun tersebut.


```python
tabel_ringkas = (
    df.groupby('tahun')
      .agg(
          total_jumlah_penderita_dm=('jumlah_penderita_dm', 'sum'),
          jumlah_kabupaten_kota=('nama_kabupaten_kota', 'nunique')
      )
      .reset_index()
)

tabel_ringkas
```

D. Soal Analisis & Visualisasi (Matplotlib)

20.	Buat grafik bar yang menampilkan jumlah_penderita_dm untuk setiap kabupaten/kota pada tahun 2019. Sumbu-x: nama_kabupaten_kota. Sumbu-y: jumlah_penderita_dm. Beri judul dan label sumbu yang jelas.


```python
import matplotlib.pyplot as plt

df_2019 = df[df['tahun'] == 2019]

plt.figure()
plt.bar(df_2019['nama_kabupaten_kota'], df_2019['jumlah_penderita_dm'])
plt.xticks(rotation=90)
plt.xlabel('Nama Kabupaten/Kota')
plt.ylabel('Jumlah Penderita DM')
plt.title('Jumlah Penderita Diabetes Mellitus per Kabupaten/Kota Tahun 2019')
plt.tight_layout()
plt.show()
```

21.	Buat grafik garis (line chart) yang menampilkan total jumlah penderita DM Jawa Barat per tahun. Sumbu-x: tahun. Sumbu-y: total jumlah_penderita_dm. Tambahkan titik data (marker) di setiap tahun.


```python
total_per_tahun = (
    df.groupby('tahun')['jumlah_penderita_dm']
      .sum()
      .reset_index()
)

plt.figure()
plt.plot(
    total_per_tahun['tahun'],
    total_per_tahun['jumlah_penderita_dm'],
    marker='o'
)

plt.xlabel('Tahun')
plt.ylabel('Total Jumlah Penderita DM')
plt.title('Total Jumlah Penderita Diabetes Mellitus di Jawa Barat per Tahun')
plt.tight_layout()
plt.show()
```

22.	Buat grafik bar horizontal untuk menampilkan 10 kabupaten/kota dengan jumlah_penderita_dm tertinggi pada tahun 2019. Urutkan dari jumlah tertinggi ke terendah.


```python
df_2019 = df[df['tahun'] == 2019]

top10_2019 = (
    df_2019.sort_values(by='jumlah_penderita_dm', ascending=False)
           .head(10)
)

plt.figure()
plt.barh(
    top10_2019['nama_kabupaten_kota'],
    top10_2019['jumlah_penderita_dm']
)
plt.xlabel('Jumlah Penderita DM')
plt.ylabel('Nama Kabupaten/Kota')
plt.title('10 Kabupaten/Kota dengan Jumlah Penderita Diabetes Mellitus Tertinggi pada Tahun 2019')
plt.gca().invert_yaxis()  # tertinggi di atas
plt.tight_layout()
plt.show()
```

23.	Buat pie chart yang menunjukkan proporsi jumlah_penderita_dm berdasarkan kategori_dm ("Rendah", "Sedang", "Tinggi") pada tahun 2019.

```python
kategori_dm_2019 = (
    df_2019.groupby('kategori_dm')['jumlah_penderita_dm'].sum()
)

plt.figure()
plt.pie(
    kategori_dm_2019,
    labels=kategori_dm_2019.index,
    autopct='%1.1f%%',
    startangle=90
)
plt.title('Proporsi Jumlah Penderita Diabetes Mellitus Berdasarkan Kategori Tahun 2019')
plt.tight_layout()
plt.show()
```

24.	Buat grafik bar yang membandingkan total jumlah_penderita_dm di tiga tahun terakhir yang tersedia (misal: 2017, 2018, 2019 jika ada di dataset).

```python
total_per_tahun = (
    df.groupby('tahun')['jumlah_penderita_dm']
      .sum()
      .reset_index()
      .sort_values('tahun')   
)

tiga_tahun_terakhir = total_per_tahun.iloc[-3:]

plt.figure()
plt.bar(
    tiga_tahun_terakhir['tahun'],
    tiga_tahun_terakhir['jumlah_penderita_dm']
)

plt.xlabel('Tahun')
plt.ylabel('Total Jumlah Penderita DM')
plt.title('Perbandingan Total Jumlah Penderita Diabetes Mellitus pada Tiga Tahun Terakhir')
plt.tight_layout()
plt.show()
```
